```python
#sort by patient number to remove duplicates
diabetes.sort(columns=['patient_nbr','encounter_id'], inplace=True)
```

```python
#drop all but first record for each patient. 
diabetes.drop_duplicates(subset=['patient_nbr'], inplace=True)
```

```python
#remove discharges to hospices or due to expiration
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 11].index, inplace=True)
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 13].index, inplace=True)
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 14].index, inplace=True)
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 19].index, inplace=True)
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 20].index, inplace=True)
diabetes.drop(diabetes[diabetes.discharge_disposition_id == 21].index, inplace=True)
```

```python
#recode three outcomes into two outcomes. Readmitted in under 30 days(1) or no(0)
def recode (admit):
    if admit=='NO':
        return 0
    if admit=='>30':
        return 0
    if admit=='<30':
        return 1
    
diabetes['outcome']=diabetes['readmitted'].map(recode)
```

```python
#recode age into float
def agerng (age):
    if age=='[0-10)':
        return 5
    if age=='[10-20)':
        return 15
    if age=='[20-30)':
        return 25
    if age=='[30-40)':
        return 35
    if age=='[40-50)':
        return 45
    if age=='[50-60)':
        return 55
    if age=='[60-70)':
        return 65
    if age=='[70-80)':
        return 75
    if age=='[80-90)':
        return 85
    if age=='[90-100)':
        return 95
    
diabetes['agecat']=diabetes['age'].map(agerng)
```

```python
diab_data = pd.DataFrame()
#save label encoders in a dictionary to have complete list. Need to save to run through test data.
label_encoders = {}

for column in diabetes.columns:
    if diabetes[column].dtype == 'object':
        #create new label encoders each time so not saving over the same label encoder. Need it for changing the test data.
        label_encoders[column] = preprocessing.LabelEncoder()
        diab_data[column] = label_encoders[column].fit_transform(diabetes[column])
    else:
        diab_data[column] = diabetes[column]
```

```python
#X_data all features except outcome. Drop arbitrary features or ones with many missing.
X_data = diab_data.drop(['encounter_id','patient_nbr','weight','payer_code','medical_specialty','readmitted','outcome'], axis=1)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

def plot_roc_curve(target_test, target_predicted_proba):
    fpr, tpr, thresholds = roc_curve(target_test, target_predicted_proba[:, 1
                                                                         ])
    roc_auc = auc(fpr, tpr)
    
    # Plot ROC curve
    plt.plot(fpr, tpr, label='ROC curve (area = %0.3f)' % roc_auc)
    plt.plot([0, 1], [0, 1], 'k--')  # random predictions curve
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate or (1 - Specifity)')
    plt.ylabel('True Positive Rate or (Sensitivity)')
    plt.title('Receiver Operating Characteristic')
    plt.legend(loc="lower right")
```

```python
y_predicted_proba = clf_lreg.predict_proba(X_test)
plot_roc_curve(y_test, y_predicted_proba)
```

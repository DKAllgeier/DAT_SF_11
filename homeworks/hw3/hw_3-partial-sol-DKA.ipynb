### Pre-Processing via Label Encoders

```python
#must to column-wise label encoding. Prevents explosion of dimensions that occurs with get_dummies
#creating a new dataframe and filling it
bank_data = pd.DataFrame()
#save label encoders in a dictionary to have complete list. Need to save to run through test data.
label_encoders = {}

for column in bank_raw.columns:
    if bank_raw[column].dtype == 'object':
        #create new label encoders each time so not saving over the same label encoder. Need it for changing the test data.
        label_encoders[column] = preprocessing.LabelEncoder()
        bank_data[column] = label_encoders[column].fit_transform(bank_raw[column])
    else:
        bank_data[column] = bank_raw[column]
```

### Train both KNN and Random Forest models

```python
# imports
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn import cross_validation
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#list comprehension
xcols = [col for col in bank_data.columns if col != 'y']

X = bank_data[xcols].values
y = bank_data['y'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
#alternative
'''x cols=[]
    for col in bank_data.columns:
        if col != 'y':
            xcols.append(col)'''
```

```python
# KNN Classifier
clf = KNeighborsClassifier()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print classification_report(y_test, y_pred)
```

```python
# Random Forest Classifier
clf = RandomForestClassifier()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

print classification_report(y_test, y_pred)
```

###But first, a protip

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing

%matplotlib inline
```

```python
cols = ['age',
        'workclass',
        'fnlwgt',
        'education',
        'education-num',
        'marital-status',
        'occupation',
        'relationship',
        'race',
        'sex',
        'capital-gain',
        'capital-loss,'
        'hours-per-week',
        'income',
        'native-country']
adult_df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data',
                       header=0,names=cols,index_col=False)
```

###Now, you're on your own.

```python
adult_df['native-country'].replace(' ?', 'unknown', inplace=True)
```

```python
adult_df['workclass'].replace(' ?', 'unknown', inplace=True)
```

```python
adult_df.income.hist()
```

```python
adult_df.income.hist(by=adult_df['workclass'],sharex=True)
```

```python
#Binarizing
binarizer = preprocessing.Binarizer(threshold=50)
binarizer.transform(adult_df.income)
```

```python
#Spliting training and test sets
X_data=adult_df[['education','workclass','race','sex','occupation','native-country']]
```

```python
#Get dummies
X_data=pd.get_dummies(X_data)
y_data=adult_df['income']
```

```python
X_data
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, random_state=12, test_size=0.2)
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = clf.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print classification_report(y_test, y_pred)
```

```python
sns.regplot(y_pred, y_test, x_jitter=0.1, y_jitter=0.1)
```

```python
from sklearn import metrics

y_pred = clf.predict(X_test)
# Confusion Matrix for Type 1 and Type 2 Error
metrics.confusion_matrix(y_test, y_pred)
```

```python
def plot_confusion_matrix(y_pred, y):
    plt.matshow(metrics.confusion_matrix(y, y_pred),
               cmap=plt.cm.binary)
    plt.colorbar()
    plt.xlabel('true value')
    plt.ylabel('predicted value')
```

```python
plot_confusion_matrix(y_pred, y_test)
```
